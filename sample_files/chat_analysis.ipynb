### 0627

```python
ds=pd.read_csv("handongsuk_0627.csv")
```

```python
goi = pd.read_csv("tmxk319_0627.csv")
```

```python
from bokeh.models import DatetimeTickFormatter
```

```python
ds
```

```python
ds_ = ds.drop_duplicates("time").reset_index(drop=True)

goi_ = goi.drop_duplicates("time").reset_index(drop=True)

ds_['viewers']=ds_['viewers'].str.replace(',','')
ds_['viewers']=ds_['viewers'].astype(int)

goi_['viewers']=goi_['viewers'].str.replace(',','')
goi_['viewers']=goi_['viewers'].astype(int)

ds_['time']=pd.to_datetime(ds_['time'])
goi_['time']=pd.to_datetime(goi_['time'])

output_notebook()

p = figure( plot_height=450, plot_width=700,title="시청자수")

p.line(x=ds_['time'],y = ds_['viewers'],color="red",legend_label='한동숙')
p.line(x=goi_['time'],y = goi_['viewers'],legend_label='괴물쥐')

p.xaxis.formatter = DatetimeTickFormatter(hourmin = ['%H:%M'])
#p.x_range.range_padding = 0.12

show(p)
```

```python
ds_ = ds.groupby('time').size().reset_index()
goi_ = goi.groupby('time').size().reset_index()
ds_=ds_.rename(columns = {0:"value"})
goi_=goi_.rename(columns = {0:"value"})
```

```python
ds_['time']=pd.to_datetime(ds_['time'])
goi_['time']=pd.to_datetime(goi_['time'])

output_notebook()

p = figure( plot_height=450, plot_width=700,title="채팅수")

p.line(x=ds_['time'],y = ds_['value'],color="red",legend_label='한동숙')
p.line(x=goi_['time'],y = goi_['value'],legend_label='괴물쥐')

p.xaxis.formatter = DatetimeTickFormatter(hourmin = ['%H:%M'])
#p.x_range.range_padding = 0.12

show(p)
```

```python
output_notebook()

p = figure(x_range = ['한동숙','괴물쥐'], plot_height=350, plot_width=700,title="채팅참여자수")

p.vbar(x=['한동숙','괴물쥐'],top = [len(ds['id'].unique()),len(goi['id'].unique())],width = 0.3,alpha=0.5,color=["red","blue"])
#p.x_range.range_padding = 0.12

show(p)
```
